## Outlier handling

# **Identifikasi Outlier dan Anomali Gunakan metode statistik atau visualisasi untuk mendeteksi outlier dalam dataset. Diskusikan bagaimana outlier dapat mempengaruhi analisis dan keputusan bisnis.**

````python
def check_plot(data, variable):
    # fungsi mengambil kerangka data (df) dan
    # variabel yang diminati sebagai argumen

    # tentukan ukuran gambar
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(data[variable], bins=30)
    plt.title('Histogram')

    # plot Q-Q
    plt.subplot(1, 3, 2)
    stats.probplot(data[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')

    # box plot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=data[variable])
    plt.title('Boxplot')

    plt.show()
````

````python
check_plot(data, 'Tenure Months')
````

````python
check_plot(data, 'Monthly Charges')
````

````python
check_plot(data, 'Total Charges')
````

````python
# Daftar kolom numerik
cols_num = ['Tenure Months', 'Monthly Charges', 'Total Charges']

# Loop untuk mendeteksi outlier
for col in cols_num: # Melakukan perulangan (loop) untuk memproses setiap kolom yang ada di list cols.
    Q1 = data[col].quantile(0.25) # Menghitung kuartil 1 (Q1) dan kuartil 3 (Q3) dari kolom:
    Q3 = data[col].quantile(0.75) # Q1 = nilai di bawah 25% data dan Q3 = nilai di bawah 75% data
    IQR = Q3 - Q1 # IQR (Interquartile Range) adalah rentang antar kuartil: Q3 dikurangi Q1.
    lower = Q1 - 1.5 * IQR # Menentukan batas bawah dan batas atas untuk mendeteksi outlier.
    upper = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower) | (data[col] > upper)] # Menyaring semua baris dari kolom yang nilainya: Lebih kecil dari lower Atau lebih besar dari upper

    print(f"Kolom: {col}")
    print(f"  Jumlah outlier: {outliers.shape[0]}") # .shape	Mengembalikan ukuran (dimensi) dari DataFrame outliers, dalam bentuk tuple: (jumlah_baris, jumlah_kolom)
    print(f"  Nilai batas bawah: {lower:.2f}, batas atas: {upper:.2f}\n") # .shape[0]	Mengambil elemen pertama dari .shape, yaitu jumlah baris (row)
````

````python
def find_outlier_boundary(data, variable):

    # Mari kita hitung batas luar yang merupakan outlier

    IQR = data[variable].quantile(0.75) - data[variable].quantile(0.25)

    lower_boundary = data[variable].quantile(0.25) - (IQR * 1.5)
    upper_boundary = data[variable].quantile(0.75) + (IQR * 1.5)

    return upper_boundary, lower_boundary
````

````python
tenure_upper_limit, tenure_lower_limit = find_outlier_boundary(data, 'Tenure Months')
tenure_upper_limit, tenure_lower_limit
````

````python
# Trimming atau drop value outlier -> pengurangan baris
# Trimming outliers from the training data
data = data[(data['Tenure Months'] <= tenure_upper_limit) & (data['Tenure Months'] >= tenure_lower_limit)]
````

# Buatlah visualiasi untuk menemukan perbedaan persentase dari customer yang churn dan customer yang tidak churn

* Pakai Histogram


````python
# Kolom target dan judul
cols = ['Churn Value', 'Churn Label']
titles = ['Persentase Churn Value', 'Persentase Churn Label']

# Subplot: 1 baris, 2 kolom
fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # Membuat dua subplot secara horizontal (1 baris, 2 kolom) dengan ukuran figure 12×5.
# axes adalah array dari dua axes (area grafik).

# Loop untuk buat bar chart persentase
for i, col in enumerate(cols): # Melakukan loop untuk setiap kolom dalam cols dan enumerate() mengembalikan indeks (i) dan nama kolom (col) sekaligus.
    # Hitung persentase
    churn_pct = data[col].value_counts(normalize=True) * 100 # normalize=True mengubah count menjadi proporsi (antara 0 dan 1), dikali 100 jadi persentase.

    # Bar chart
    sns.barplot(x=churn_pct.index, y=churn_pct.values, ax=axes[i]) # Menggunakan axes[i] untuk menempatkan plot di subplot ke-i
    axes[i].set_title(titles[i]) # Memberi judul pada masing-masing subplot dari list titles.
    axes[i].set_ylabel('Persentase (%)') # Memberi label sumbu Y.
    axes[i].set_ylim(0, 100)  # Mengatur skala sumbu Y dari 0 hingga 100 agar konsisten antar plot dan mudah dibandingkan.
    for j, val in enumerate(churn_pct.values): # Melakukan loop ke setiap nilai persentase (val) pada hasil bar chart. j adalah posisi indeks dari batangnya.
        axes[i].text(j, val + 1, f'{val:.2f}%', ha='center')  # axes[i].text(...)	Menambahkan teks ke subplot ke-i (karena kita sedang plotting di axes[i]).
# j	Posisi X dari teks. Harus cocok dengan posisi batang ke-j agar teks muncul tepat di atas batangnya.
# val + 1	Posisi Y dari teks. Tambah 1 agar teks tidak menempel langsung ke atas batang, supaya lebih enak dilihat
# f'{val:.2f}%'	Format teks: ambil val, bulatkan ke 2 desimal, lalu tambahkan tanda %. Contoh: 74.52%.
# ha='center'	Horizontal alignment = center → teks disejajarkan secara tengah dengan batang.
plt.tight_layout()
plt.show()
````

# Buatlah visualisasi untuk melihat 5 kota teratas yang customernya melakukan churn dan tidak melakukan churn


````python
# 1. Hitung jumlah churn & non-churn per kota
top_cities = data.groupby(['City', 'Churn Value']).size().reset_index(name='Count')
# groupby(['City', 'Churn Value'])	Mengelompokkan data berdasarkan nama kota dan status churn (0 = tidak churn, 1 = churn)
# .size()	Menghitung jumlah baris (customer) di setiap grup
# .reset_index(name='Count')	Mengubah hasil groupby menjadi DataFrame baru dan beri nama kolom jumlahnya “Count”

# 2. Ambil total churn + non-churn, lalu ambil 5 kota dengan jumlah customer terbanyak
top_5_city_names = (
    top_cities.groupby('City')['Count'] # groupby('City')['Count'].sum()	Menjumlahkan seluruh customer (churn dan tidak churn) per kota
    .sum()
    .sort_values(ascending=False) # .sort_values(ascending=False)	Urutkan kota dari jumlah customer tertinggi ke terendah
    .head(5) # .head(5)	Ambil hanya 5 kota teratas
    .index # .index	Ambil nama-nama kotanya sebagai index (untuk dipakai filter nanti)
)

# 3. Filter hanya 5 kota teratas
top_cities_filtered = top_cities[top_cities['City'].isin(top_5_city_names)]
# top_cities[...]	Ambil baris dari data awal (yang punya info churn & count)
# .isin(top_5_city_names)	Hanya pilih baris dengan nama kota yang ada di daftar top 5 kota tadi

# 4. Plot
plt.figure(figsize=(10, 6)) # plt.figure(figsize=(10, 6))	Membuat canvas plot ukuran 10x6 inci
ax = sns.barplot( # sns.barplot(...)	Membuat barplot
    data=top_cities_filtered,
    x='City', # x='City'	Sumbu X adalah nama kota
    y='Count', # y='Count'	Sumbu Y adalah jumlah customer
    hue='Churn Value',  # hue='Churn Value'	Memisahkan warna bar berdasarkan apakah customer churn (1) atau tidak (0)
    palette='Set2' # palette='Set2'	Mengatur warna supaya beda antara churn dan tidak churn
)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=3)
plt.title('Top 5 Kota dengan Customer Churn vs Tidak Churn') #Menambahkan judul pada grafik
plt.xlabel('Kota')
plt.ylabel('Jumlah Customer') # Label untuk sumbu X dan Y
handles, labels = ax.get_legend_handles_labels()# Memberi nama legend (warna) di samping
new_labels = ['0 = Tidak Churn', '1 = Churn']
ax.legend(handles=handles, labels=new_labels, title='Status Churn', loc='upper right')
plt.grid(axis='y') #Menambahkan garis bantu horizontal agar nilai lebih mudah dibaca
plt.tight_layout() # Mengatur layout agar tidak tabrakan
plt.show()
````
